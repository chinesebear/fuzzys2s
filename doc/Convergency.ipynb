## 统计和收敛
EUconst、HS、XLSum三个数据集

### read data

```python
import csv
import numpy as np
base_path="/home/yang/github/fuzzys2s/"
euconst_path = base_path+'output/train_transformer_opus_euconst.csv'
euconst_path2 = base_path+'output/train_fuzzys2s_opus_euconst.csv'
hs_path = base_path+'output/train_transformer_hearthstone.csv'
hs_path2 = base_path+'output/train_fuzzys2s_hearthstone.csv'
xlsum_path = base_path+'output/train_transformer_xlsum.csv'
xlsum_path2 = base_path+'output/train_fuzzys2s_xlsum.csv'
euconst_data = []
hs_data = []
xlsum_data = []
with open(euconst_path, encoding='utf-8') as f:
    reader = csv.reader(f)
    headers = next(reader)
    print(headers)
    for row in reader:
        euconst_data.append([float(row[0]), float(row[1])])

with open(euconst_path2, encoding='utf-8') as f:
    reader = csv.reader(f)
    headers = next(reader)
    print(headers)
    for row in reader:
        euconst_data.append([float(row[0]), float(row[1])])

with open(hs_path, encoding='utf-8') as f:
    reader = csv.reader(f)
    headers = next(reader)
    print(headers)
    for row in reader:
        hs_data.append([float(row[0]), float(row[1])])

with open(hs_path2, encoding='utf-8') as f:
    reader = csv.reader(f)
    headers = next(reader)
    print(headers)
    for row in reader:
        hs_data.append([float(row[0]), float(row[1])])

with open(xlsum_path, encoding='utf-8') as f:
    reader = csv.reader(f)
    headers = next(reader)
    print(headers)
    for row in reader:
        xlsum_data.append([float(row[0]), float(row[1])])
        
# with open(xlsum_path2, encoding='utf-8') as f:
#     reader = csv.reader(f)
#     headers = next(reader)
#     print(headers)
#     for row in reader:
#         xlsum_data.append([float(row[0]), float(row[1])])

def sampling(data, step=10):
    output=[]
    for i in range(len(data)):
        if i % step == 0:
            output.append(data[i])
    return output

```

### plot

```python
from matplotlib import pyplot as plt
import numpy as np

euconst_data = np.array(sampling(euconst_data))
hs_data = np.array(sampling(hs_data))
xlsum_data = np.array(sampling(xlsum_data))

f, axes = plt.subplots(2, 3, figsize=(15,10))

sub_graph = axes[0][0]
data = euconst_data
sub_graph.plot(data[:, 0], label='Loss',color='r')
sub_graph.set_title("EUconst(FuzzyS2S-T)")
sub_graph.set_xlabel("Epoch")
sub_graph.set_ylabel("Loss")
sub_graph.legend()

sub_graph = axes[0][1]
data = hs_data
sub_graph.plot(data[:, 0], label='Loss',color='r')
sub_graph.set_title("HS(FuzzyS2S-T)")
sub_graph.set_xlabel("Epoch")
sub_graph.set_ylabel("Loss")
sub_graph.legend()

sub_graph = axes[0][2]
data = xlsum_data
sub_graph.plot(data[:, 0], label='Loss',color='r')
sub_graph.set_title("XLSum(FuzzyS2S-T)")
sub_graph.set_xlabel("Epoch")
sub_graph.set_ylabel("Loss")
sub_graph.legend()

sub_graph = axes[1][0]
data = euconst_data
sub_graph.plot(data[:, 1], label='Valid Acc',color='b')
sub_graph.set_title("EUconst(FuzzyS2S-T)")
sub_graph.set_xlabel("Epoch")
sub_graph.set_ylabel("Accuracy")
sub_graph.legend()

sub_graph = axes[1][1]
data = hs_data
sub_graph.plot(data[:, 1], label='Valid Acc',color='b')
sub_graph.set_title("HS(FuzzyS2S-T)")
sub_graph.set_xlabel("Epoch")
sub_graph.set_ylabel("Accuracy")
sub_graph.legend()

sub_graph = axes[1][2]
data = xlsum_data
sub_graph.plot(data[:, 1], label='Valid Acc',color='b')
sub_graph.set_title("XLSum(FuzzyS2S-T)")
sub_graph.set_xlabel("Epoch")
sub_graph.set_ylabel("Accuracy")
sub_graph.legend()
```
